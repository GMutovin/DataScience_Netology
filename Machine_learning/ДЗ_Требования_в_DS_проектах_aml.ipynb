1. Как бы Вы организовали аналитику по данному продукту?

Выбор гибридной модели организации аналитики с выделением команды аналитиков под продукт.

Описание процесса работы с данными:
1. Описание событий входа в процесс
2. Описание этапа анализа бизнес потребности и этапа принятия решения о необходимости анализа данных.
3. Описание этапа формирования задач для аналитиков от владельца продукта.ё
4. Описание этапа по получению необходимых данных.
5. Этап по анализу данных командой аналитиков. 
6. Этап подготовки результата работы.
7. Предоставление результата заказчику.

Настройка интеграции внутренних систем с сайтом, где продукт реализуется, для сбора данных.

Организация хранения собранных данных (включая архитектуру и обеспечение доступа к данным) и структуирования данных в БД.

Организация сбора данных о динамике прохождения курсов, количестве выполненных домашних заданий, удовлетворенности курсами, проценту возврата студентов к отдельным темам, количеству активных студентов, посещаемости вебинаров, времени на просмотр лекции/вебинара и отдельно по паузам в рамках отдельной лекции/вебинара, cookies.

Организация процесса выгрузок данных.



2. Сформулируйте 2-3 ключевые метрики продукта

Реферальность - доля новых учеников, которые пришли по рекомендации текущего. Поможет отследить коммерческую ценность продукта. 

CSI - удовлетворенность студентов контентом и занятиями. Опросы студентов по качеству материала и сложности отдельных тем.

NPS - лояльность студентов и готовность рекомендовать продукт.

COR - доля студентов, завершивших курсы. 


3. Какие данные Вам могут помочь в поиске инсайтов?

Скорость проверки домашних заданий и динамика обратной связи. Поможет отследить необходимость привлечения дополнительных специалистов для работы над продуктом.

Уровень сложности курсов: 
1. Доля студентов, успешно завершивших курсы и получивших диплом. Поможет отследить качество предоставляемого продукта. Также необходимо отслеживать прогресс студентов для выявления сложных тем, чтобы улучшить конкретный курс.
2. Доля студентов, которые успешно трудоутроились после завершения курсов. Поможет также отследить недостатки курсов.
3. Время, которое затрачивается студентами в среднем для освоения темы.
4. Среднее время, затрачиваемое для просмотра лекции/семинара.
5. Среднее время для выполнения домашних заданий и количество успешно выполненных домашних заданий.
6. Динамика посещения занятий.

Также ценными данными являются показатели обратной связи студентов.


4. Опишите команду для работы с данными для данного продукта. Какие роли Вам будут нужны? Чем они будут заниматься?

* Владелец продукта для организации работы команды, трансформации бизнес требований заказчиков в задачи для аналитиков. 

* Разработчики для настройки системы для сбора данных.  

* Дата инженеры для подготовки данных для дальнейшей аналитики.

* Продуктовые аналитики для анализа метрик продукта и проведения АB тестов, предоставления рекомендаций бизнесу. Дата саенс специалисты для построения аналитических моделей и извлечения инсайтов.

* Специалисты по БД для настройки оптимального хранения данных и доступа к ним.
